```python
import os
import matplotlib.pyploy as plt
```

```python
base_dir="./data"
for folder in os.listdir(base_dir):
    folder_path=os.path.join(base_dir,folder)
    first_three_images=[os.path.join(folder_path,image) for image in os.listdir(folder_path)[:3]]
    plt.Figure(figsize=(12,4))
    plt.suptitle(f"{folder} {len(os.listdir(folder_path))} images")
    for i,image in enumerate(first_three_images,1):
        plt.subplot(1,3,i)
        plt.imshow(plt.imread(image))

    plt.show()
```
